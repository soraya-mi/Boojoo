```python
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
import matplotlib.pyplot as plt
dataframe = pd.read_csv("groceries.csv", header=None)
```

```python
records = []
for i in range (0, 9835):
    records.append([str(dataframe.values[i,j]) for j in range(0, 32)])
for i in range(0,4):
  print(records[i])
items=[]
f_records=[]
for i in records:
  i=[x for x in i if x != 'nan']
  f_records.append(i)
  for j in i:
    if not items.__contains__(j):
      items.append(j)
print(items)
print(len(items))
# for i in range(0,30):
#   print(f_records[i])
```

```python
te = TransactionEncoder()
te_ary = te.fit(f_records).transform(f_records)
# te_ary.astype("int")
df=pd.DataFrame(te_ary, columns=te.columns_)
  
```
